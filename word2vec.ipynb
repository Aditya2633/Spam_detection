```python
import pandas as pd
messages = pd.read_csv("C:\\Users\\91772\\my_nltk_project\\spam.csv", encoding='latin1')
```

```python
messages.rename(columns= {'v1' : 'lable', 'v2':'message'} , inplace=1)
```

```python
messages.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1, inplace=True)
```

```python
messages.head()
```

```python
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus=[]

for i in range(0 , len(messages)):
    review = re.sub('[^a-z A-Z]' ,' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

create bag of words


```python
import numpy as np
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500 , ngram_range=(1,2))
```

```python
x = cv.fit_transform(corpus).toarray()
```

```python
x
```

```python
cv.vocabulary_
```

```python
y = pd.get_dummies(messages['lable'])
```

```python
y=y.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train ,y_test = train_test_split(x,y , test_size=0.20 )
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect = MultinomialNB().fit(x_train,y_train)
```

```python
y_pred = spam_detect.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score , classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
